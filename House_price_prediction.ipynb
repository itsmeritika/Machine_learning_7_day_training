```python
#Step 1: Data Collection
import pandas as pd
```

```python
data = pd.read_csv('housing.csv')
```

```python
#Step 2: Data Cleaning and Preprocessing
data.head()
```

```python
data.drop(columns=['longitude','latitude','housing_median_age','total_bedrooms','households'],inplace=True)
```

```python
data.head()
```

```python
#Step 3: Data Splitting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(data.drop(columns=['median_house_value']),data['median_house_value'],test_size=0.2)
```

```python
data.drop(columns=['ocean_proximity'],inplace=True)
```

```python
#Step 4: Model Selection and Training
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.predict([[1627,565,3.84]])
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
y_pred = model.predict(x_train)
mean_squared_error(y_train,y_pred)
```

```python
mean_absolute_error(y_train,y_pred)
```

```python
from math import sqrt
sqrt(mean_squared_error(y_train,y_pred))
```
