```python
import pandas as pd
movies = pd.read_csv("movies.csv")
movies.head()

```

```python
rating = pd.read_csv("ratings.csv")
rating.head()
```

```python
final_data = rating.pivot(index='movieId',columns='userId', values='rating')
final_data.head()
```

```python
final_data.fillna(0,inplace=True)
final_data.head()
```

```python
no_user_voted = ratings.groupby('movieId')['rating'].agg('count')
no_user_voted
```

```python
no_movies_voted = ratings.groupby('userId')['rating'].agg('count')
no_movies_voted
```

```python
no_user_voted = ratings.groupby('movieId')['rating'].agg('count')
no_user_voted
```

```python
import matplotlib.pyplot as plt
plt.scatter(no_user_voted.index,no_user_voted)
plt.xlabel('MovieId')
plt.ylabel('No of users voted')
plt.show()
```

```python
final_dataset = final_data.loc[no_user_voted[no_user_voted>10].index,:]
final_dataset
```

```python
final_dataset = final_dataset.loc[:,no_movies_voted[no_movies_voted>30].index]
final_dataset
```

```python
import numpy as np
sample = np.array([[0,0,2,0],[0,1,0,0],[0,0,0,4]])
sparsity = 1.0 - (np.count_nonzero(sample)/float(sample.size))
print(sparsity)
```

```python
from scipy.sparse import csr_matrix
csr_sample = csr_matrix(sample)
print(csr_sample)
```

```python
csr_dataset = csr_matrix(final_dataset.values)
print(csr_dataset)
```

```python
from sklearn.neighbors import NearestNeighbors
knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=20)
knn.fit(csr_dataset)
```

```python
final_dataset.reset_index(inplace=True)
final_dataset.head()
```

```python
def get_movie_recommendation(movie_name, n_movies_to_recom=10):

  movie_list = movies[movies['title'].str.contains(movie_name)]

  if len(movie_list):
    movie_idx = movie_list.iloc[0]['movieId']
    movie_idx = final_dataset[final_dataset['movieId']==movie_idx].index[0]
    distances, indices = knn.kneighbors(csr_dataset[movie_idx],n_neighbors = n_movies_to_recom+10)
    rec_movie_indices = sorted(list(zip(indices.squeeze().tolist(),distances.squeeze().tolist())),key=lambda x:x[1])
    recommend_frame = []
    for val in rec_movie_indices:
       idx = movies[movies['movieId']==val[0]]
       recommend_frame.append({'Title':movies.iloc[idx]['title'].values[0],'Distance':val[1]})
       df = pd.Dataframe(recommend_frame,index=range(1,n_movies_to_recom+1))
       return df

    else:
      return "No movies found.please check the name of your movie"



   # n_movies_to_reccomend = 10
    #movie_list = movies[movies['title'].str.contains(movie_name)]
```

```python
def get_movie_recommendation(movie_name,n_movies_to_reccomend = 10):

    movie_list = movies[movies['title'].str.contains(movie_name)]
    if len(movie_list):
        movie_idx= movie_list.iloc[0]['movieId']
        movie_idx = final_dataset[final_dataset['movieId'] == movie_idx].index[0]
        distances , indices = knn.kneighbors(csr_dataset[movie_idx],n_neighbors=n_movies_to_reccomend+1)
        rec_movie_indices = sorted(list(zip(indices.squeeze().tolist(),distances.squeeze().tolist())),key=lambda x: x[1])[1:]
        recommend_frame = []
        for val in rec_movie_indices:
            movie_idx = final_dataset.iloc[val[0]]['movieId']
            idx = movies[movies['movieId'] == movie_idx].index
            recommend_frame.append({'Title':movies.iloc[idx]['title'].values[0],'Distance':val[1]})
        df = pd.DataFrame(recommend_frame,index=range(1,n_movies_to_reccomend+1))
        return df
    else:
        return "No movies found. Please check your input"
```
